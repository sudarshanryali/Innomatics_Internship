$$Histogram$$

* A histogram is a graphical display of data using bars of different heights. In a histogram, each bar groups numbers into ranges. Taller bars show that more data falls in that range. A histogram displays the shape and spread of continuous sample data.

```python
data[['conscientiousness','agreeableness','extraversion','nueroticism','openess_to_experience']].hist(figsize=(12,9))
```

$$BoxPlot$$

* A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”). It can tell you about your outliers and what their values are

```python
def boxplot(column,dataf):
    plt.figure(figsize=(10,4))
    sns.boxplot(x=column,data=dataf)
```

```python
nums=['conscientiousness','agreeableness','extraversion','nueroticism','openess_to_experience']
for i in nums:
    boxplot(i,data)
```

```python
cat_list=['ComputerScience','MechanicalEngg','ElectricalEngg','TelecomEngg','CivilEngg']
for col in cat_list:
    plt.figure(figsize=(8,5))
    sns.boxplot(x=col,data=data)
```

$$Kernel Density Plot$$

* A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analagous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.


```python
cat_list=['conscientiousness','agreeableness','extraversion','nueroticism','openess_to_experience']
for col in cat_list:
    plt.figure(figsize=(8,5))
    sns.kdeplot(x=col,data=data)
```

$$ Cummilative   distribution   Function$$

* The PMF is one way to describe the distribution of a discrete random variable. As we will see later on, PMF cannot be defined for continuous random variables. The cumulative distribution function (CDF) of a random variable is another method to describe the distribution of random variables. The advantage of the CDF is that it can be defined for any kind of random variable (discrete, continuous, and mixed).

```python
cat_list=['conscientiousness','agreeableness','extraversion','nueroticism','openess_to_experience']
for col in cat_list:
    plt.figure(figsize=(8,5))
    sns.ecdfplot(x=col,data=data)
```

```python
cat_list=['ComputerScience','MechanicalEngg','ElectricalEngg','TelecomEngg','CivilEngg']
for col in cat_list:
    plt.figure(figsize=(8,5))
    sns.kdeplot(x=col,data=data)
```

$$Violin Plot$$

* A violin plot is a method of plotting numeric data. It is similar to a box plot, with the addition of a rotated kernel density plot on each side. ... While a box plot only shows summary statistics such as mean/median and interquartile ranges, the violin plot shows the full distribution of the data.

```python
cat_list=['ComputerScience','MechanicalEngg','ElectricalEngg','TelecomEngg','CivilEngg']
for col in cat_list:
    plt.figure(figsize=(8,5))
    sns.set_theme(style="whitegrid")
    sns.violinplot(x=col,data=data)
```

```python
cat_list=['conscientiousness','agreeableness','extraversion','nueroticism','openess_to_experience']
for col in cat_list:
    plt.figure(figsize=(8,5))
    sns.set_theme(style="whitegrid")
    sns.violinplot(x=col,data=data)
```
